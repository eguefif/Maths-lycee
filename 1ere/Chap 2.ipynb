```python
# Chap 2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ft_abs(x):
    if x < 0:
        x = -x
    return x

x = np.arange(-10, 10, 0.1)
y = np.array([ft_abs(a) for a in x])

plt.plot(x, y)
```

```python
x = np.arange(-10, 10, 0.1)
y = np.array([2*(2-a)*(-a-1) for a in x])

plt.plot(x, y)
```

```python
# Exercice 61
def f(x):
    return x**2/7 -x*5**(1/2)+3

def abs(x):
    return (x**2)**(1/2)

x = np.arange(-20, 20, 0.0001)
y = [f(a) for a in x]

for k, a in enumerate(y):
    if k+1 >= len(y):
        break
    if a*y[k+1] < 0:
        if a > y[k+1]:
            print(f"For x={x[k+1]} f({x[k+1]}) = 0.")
        else:
            print(f"For x={x[k+1]} f({x[k+1]}) = 0.")
```
